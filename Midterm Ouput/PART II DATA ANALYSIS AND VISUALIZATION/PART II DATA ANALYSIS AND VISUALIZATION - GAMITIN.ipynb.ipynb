```python
import pandas as pd
import hvplot.pandas
from bokeh.models.formatters import BasicTickFormatter #For converting scientific notation to integers 

#PHILIPPINES DENGUE CASES 2016-2020
#https://www.kaggle.com/datasets/vincentgupo/dengue-cases-in-the-philippines

dataset = pd.read_csv('ph_dengue_cases2016-2020.csv')

#DATA ANALYSIS QUESTIONS
#1. What is the number of Dengue Cases per year?
#2. What is the total number of Dengue Cases in each Region?
#3. What is the total Dangue Cases and Dengue Deaths per Year?
#4. What is the total Dengue cases of each Region in every Year?
#5. What are the Average Dengue Cases per Month? 
```

```python
#1. What is the number of Dengue Cases per year?

year = dataset[['Year', 'Dengue_Cases']].groupby(['Year']).sum()

year
```

```python
#2. What is the total number of Dengue Cases in each Region?

region = dataset[['Region', 'Dengue_Cases']].groupby(['Region'], sort=False).sum()

region
```

```python
#3. What is the total Dangue Cases and Dengue Deaths per Year?

deaths = dataset[['Year', 'Dengue_Cases', 'Dengue_Deaths']].groupby(['Year'], sort=False).sum()

deaths
```

```python
#4. What is the total Dengue cases of each Region in every Year?

region_year = dataset.groupby(by=['Year', 'Region'], sort=False).agg({'Dengue_Cases': 'sum'}).sort_values(by='Year')

region_year
```

```python
#5. What are the Average Dengue Cases per Month? 

dengue_month = dataset.groupby(by=['Month'], sort=False).agg({'Dengue_Cases': 'mean'}).round(2)

dengue_month
```

```python
#1. What is the number of Dengue Cases per year?

year.hvplot(title='Number of Dengue Cases per Year', yformatter=BasicTickFormatter(use_scientific=False))
```

```python
#2.What is the total number of Dengue Cases in each Region?

region.hvplot.bar(title='Total Number of Dengue Cases in Each Region', rot=90, yformatter=BasicTickFormatter(use_scientific=False))
```

```python
#3. What is the total Dangue Cases and Dengue Deaths per Year?

deaths.hvplot.step(title='Total Dangue Cases and Dengue Deaths per Year', height=600, rot=90, yformatter=BasicTickFormatter(use_scientific=False))
```

```python
#4. What is the total Dengue cases of each Region in every Year?

region_year.hvplot.bar(title='Dengue cases of each Region in every Year', width=2500, rot=90, yformatter=BasicTickFormatter(use_scientific=False))
```

```python
#5. What are the Average Dengue Cases per Month? 

dengue_month.hvplot.area(title='Average Dengue Cases per Month', color='red', width=1000,yformatter=BasicTickFormatter(use_scientific=False))
```
